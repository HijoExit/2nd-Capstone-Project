```python
import pandas as pd
import requests
import numpy as np
import matplotlib as plt 
```

```python
#CSV data files
gdp = pd.read_csv('gdp.CSV')
gdp_growth = pd.read_csv('gdp_growth.csv')
gpd_per_capita = pd.read_csv('gdp_per_capita.csv')
gdp_per_capita_growth = pd.read_csv('gdp_per_capita_growth.csv')
gni = pd.read_csv('API_NY.GNP.PCAP.CD_DS2_en_csv_v2_4250945.csv')
Chinese_pop = pd.read_csv('Chinas Population En.csv')
Japanese_pop =pd.read_csv('japanese_census.csv')
Fertility_rate = pd.read_csv('API_SP.DYN.TFRT.IN_DS2_EN_csv_v2_4023233.csv')
Age_14under = pd.read_csv('API_SP.POP.0014.TO.ZS_DS2_en_csv_v2_4153151.csv')
Age_dependency= pd.read_csv('API_SP.POP.DPND_DS2_en_csv_v2_4154688.csv')
Age_15to64 = pd.read_csv('API_SP.POP.1564.TO.ZS_DS2_en_csv_v2_4150831.csv')
Age_64above = pd.read_csv('API_SP.POP.65UP.TO.ZS_DS2_en_csv_v2_4154737.csv')
Meta_data = pd.read_csv('Metadata_Country_API_SP.POP.DPND_DS2_en_csv_v2_4154688.csv')
Life =pd.read_csv('API_SP.DYN.LE00.IN_DS2_en_csv_v2_4250211.csv')
Birth_rate = pd.read_csv('API_SP.DYN.CBRT.IN_DS2_en_csv_v2_4252397.csv')
Death_rate = pd.read_csv('API_SP.DYN.CDRT.IN_DS2_en_csv_v2_4252395.csv')
Female_pop = pd.read_csv('API_SP.POP.TOTL.FE.ZS_DS2_en_csv_v2_4262603.csv')
Total_pop =pd.read_csv('API_SP.POP.TOTL_DS2_en_csv_v2_4251154.csv')
Pop_growth = pd.read_csv('API_SP.POP.GROW_DS2_en_csv_v2_4251293.csv')
GNI = pd.read_csv('API_NY.GNP.PCAP.CD_DS2_en_csv_v2_4250945.csv')
Trade_balance = pd.read_csv('API_NE.RSB.GNFS.CD_DS2_en_csv_v2_4261356.csv')
Gov_exp = pd.read_csv('API_NE.CON.GOVT.CD_DS2_en_csv_v2_4261177.csv')
```

```python
def demographic_data_clean(x):
    header = x.iloc[3:4]
    x.columns = header.iloc[0]
    df= x.drop(columns = ['Country Code', 'Indicator Name', 'Indicator Code'])
    df2 = df.iloc[4:].set_index('Country Name').T
    df3 = df2.reset_index().rename(columns = {3:'Year'})
    df3['Year'] = df3['Year'].astype(int)
    return df3.set_index('Year')
def Economic_data_clean(x):
    df= x.drop(columns = 'Code')
    df2 = df.set_index('Country Name').T
    df3 = df2.reset_index().rename(columns = {'index':'Year'})
    df4 = df3.iloc[0:-1].set_index('Year')
    df4.index = df4.index.astype(int)
    return df4
```

```python
#Cleaning up the Fertility rate 
F_rate = Fertility_rate[3:] 
F_rate.columns = F_rate.iloc[0]
F = F_rate[1:-1]
F.set_index('Country Name',inplace = True)
F_rate_CN_JP  = F.loc[['China', 'Japan', 'Korea, Rep.', 'Singapore', 'Malaysia']]
F_rate_CN_JP.drop(columns = ['Indicator Name', 'Indicator Code'], inplace = True)
F_rate_CN_JP_T = F_rate_CN_JP.T
F_rate_CN_JP_T.reset_index(inplace = True)
F_rate_CN_JP_T.columns = ["Year", 'China', \
                           'Japan',\
                          'Korea, Rep.', \
                          'Singapore',\
                          'Malaysia' ]
F_rate_CN_JP_T1 = F_rate_CN_JP_T.loc[1:]
F_rate_CN_JP_T1['Year'] = F_rate_CN_JP_T1['Year'].astype(int)
F_rateDF = F_rate_CN_JP_T1.set_index('Year')
```

```python
#Creating df for different features we will be using later 
BirthDF= demographic_data_clean(Birth_rate)
Age_14under_DF = demographic_data_clean(Age_14under)
Age_dependency_DF= demographic_data_clean(Age_dependency)
Age_15to64_DF = demographic_data_clean(Age_15to64)
Age_64above_DF = demographic_data_clean(Age_64above)
Life_DF = demographic_data_clean(Life)
Death_df = demographic_data_clean(Death_rate)
Female_pop_df= demographic_data_clean(Female_pop)
Total_pop_df = demographic_data_clean(Total_pop)
pop_growth = demographic_data_clean(Pop_growth)
GNI_df = demographic_data_clean(GNI)
Trade_df = demographic_data_clean(Trade_balance)
Gov_df = demographic_data_clean(Gov_exp)
gdp_df = Economic_data_clean(gdp)
gdp_df
```

```python
Japan =  pd.DataFrame()
China =  pd.DataFrame()
Korea =  pd.DataFrame()
Singapore =  pd.DataFrame()
Malaysia  =  pd.DataFrame()
def country_df_creation(x,y):
    x['GNI per Capita'] = GNI_df[y]
    x['14 Under'] = Age_14under_DF[y]
    x['15 to 64'] = Age_15to64_DF[y]
    x['64 Above'] = Age_64above_DF[y]
    x['Female(%)'] = Female_pop_df[y]
    x['Total pop'] = Total_pop_df[y]
    x['Life expectancy at birth'] =Life_DF[y]
    x['Fertility rate, total (births per woman)'] = F_rateDF[y]
    x['Total pop'] = Total_pop_df[y]
    x['Pop Growth'] = pop_growth[y]
    x['Work Population(%)'] = Age_dependency_DF[y]
    x['Death rate']  = Death_df[y]
    x['Birth rate'] = BirthDF[y]
    x['Trade balance'] = Trade_df[y]
    x['Gov Exp'] = Gov_df[y]
    return x
Japan = country_df_creation(Japan, 'Japan')
China = country_df_creation(China, 'China')
Korea = country_df_creation(Korea, 'Korea, Rep.')
Singapore = country_df_creation(Singapore, 'Singapore')
Malaysia = country_df_creation(Malaysia, 'Malaysia')
Japan['GDP']  = gdp_df['Japan']
China['GDP']  = gdp_df['China']
Korea['GDP']  = gdp_df['Korea, Rep.']
Singapore['GDP']  = gdp_df['Singapore']
Malaysia['GDP']  = gdp_df['Malaysia']
Japan.to_csv('Japan.csv')
China.to_csv('China.csv')
Korea.to_csv('Korea.csv')
Singapore.to_csv('Singapore.csv')
Malaysia.to_csv('Malaysia.csv')
```

```python
Meta_data.set_index('IncomeGroup', inplace = True)
```

```python
#Separating out highincome and middleincome countries
Highincome = Meta_data.drop(['Lower middle income', 'Low income', 'Upper middle income', np.nan ])
UpperMidincome =   Meta_data.drop(['Lower middle income', 'Low income', 'High income', np.nan ])
```

```python
#Renaming the code to match the datasets laer. 
Highincome.rename(columns = {'Country Code':'Code'}, inplace = True)
UpperMidincome.rename(columns = {'Country Code':'Code'}, inplace = True)
F.rename(columns = {'Country Code':'Code'}, inplace = True)
Age_dependency.rename(columns = {'Country Code':'Code'}, inplace = True)
Age_14under.rename(columns = {'Country Code':'Code'}, inplace = True)
Age_dependency.rename(columns = {'Country Code':'Code'}, inplace = True)
Age_15to64.rename(columns = {'Country Code':'Code'}, inplace = True)
Age_64above.rename(columns = {'Country Code':'Code'}, inplace = True)
```

```python
#Creating function to create a dataframe for different categories
def Function(x,y):
    DF1 = x[x.Code.isin(y.Code)]
    DF2 = DF1.set_index('Country Name').reset_index().T
    DF3 = DF2.iloc[2:-1]
    DF4 = DF3.reset_index().rename(columns = {'index':'Year'})
    DF4['Year'] = DF4['Year'].astype(int)
    DF4.set_index('Year', inplace = True)
    thresh = len(DF4) * .9
    DF4.dropna(thresh = thresh, axis = 1, inplace = True)
    return DF4
def Function2(x,y):
    DF1 = x[x.Code.isin(y.Code)]
    Df2 = DF1.iloc[4:].T
    Df3 = Df2.reset_index()
    Df4 = Df3.rename(columns = {3:'Year'})
    Df5 = Df4.loc[4:]
    Df5['Year'] = Df5['Year'].astype(int)
    Df6 = Df5.loc[4:].set_index('Year')
    thresh = len(Df6) * .2
    Df6.dropna(thresh = thresh, axis = 1, inplace = True)
    return Df6
```

```python
#Highincome country data
Highincome65 = Function2(Age_64above, Highincome)
Highincome14 =Function2(Age_14under, Highincome)
Highincome15to64 = Function2(Age_15to64, Highincome)
HighincomeF_rate = Function2(F, Highincome)
Highincome_Age_dependency = Function2(Age_dependency, Highincome)
Highincome_GDP_Growth = Function(gdp_growth, Highincome)
Highincome_GDP_per_capita = Function(gpd_per_capita, Highincome)
Highincome_GDP_per_capita_growth = Function(gdp_per_capita_growth, Highincome)
Highincome_DF = pd.DataFrame()
Highincome_DF['GDP Growth'] = Highincome_GDP_Growth.mean(axis = 1)
Highincome_DF['GDP Per Capita']=Highincome_GDP_per_capita.mean(axis = 1)
Highincome_DF['14 Under'] = Highincome14.mean(axis = 1)
Highincome_DF['64 Above'] = Highincome65.mean(axis = 1)
Highincome_DF['15 to 64'] = Highincome15to64.mean(axis = 1)
Highincome_DF['Fertility rate, total (births per woman)'] = HighincomeF_rate.mean(axis = 1)
Highincome_DF['Work Population(%)'] = Highincome_Age_dependency.mean(axis = 1)
Highincome_DF.to_csv('Highincome_DF.csv')
```

```python
#Highincome country data
UpperMidincome65 = Function2(Age_64above, UpperMidincome)
UpperMidincome14 =Function2(Age_14under, UpperMidincome)
UpperMidincome15to64 = Function2(Age_15to64, UpperMidincome)
UpperMidincomeF_rate = Function2(F, UpperMidincome)
UpperMidincome_GDP_Growth = Function(gdp_growth, UpperMidincome)
UpperMidincome_Age_dependency = Function2(Age_dependency, UpperMidincome)
UpperMidincome_Growth = Function(gdp_growth, UpperMidincome)
UpperMidincome_per_capita = Function(gpd_per_capita, UpperMidincome)
UpperMidincome_per_capita_growth = Function(gdp_per_capita_growth, UpperMidincome)
UpperMidincome_DF = pd.DataFrame()
UpperMidincome_DF['GDP Growth'] = UpperMidincome_GDP_Growth.mean(axis = 1)
UpperMidincome_DF['GDP Per Capita']=UpperMidincome_per_capita.mean(axis = 1)
UpperMidincome_DF['14 Under'] = UpperMidincome14.mean(axis = 1)
UpperMidincome_DF['64 Above'] = UpperMidincome65.mean(axis = 1)
UpperMidincome_DF['15 to 64'] = UpperMidincome15to64.mean(axis = 1)
UpperMidincome_DF['Fertility rate, total (births per woman)'] = UpperMidincomeF_rate.mean(axis = 1)
UpperMidincome_DF['Work Population(%)'] = UpperMidincome_Age_dependency.mean(axis = 1)
UpperMidincome_DF.to_csv('UpperMidincome_DF.csv')
```
